# 機率(Probability)

## 排列(permutation)與組合(Combination)

## 範例1. 有三個球排列的事件共有幾種?

```python
from itertools import permutations 

# 測試資料
list1 = [1, 2, 3]

# 宣告排列的類別
perm = permutations(list1) 

# 印出所有事件
print('所有事件:')
list_output = list(perm)
for i in list_output: 
    print(i) 
    
print() 
print(f'排列數={len(list_output)}')
```

## 範例2. 有三個球，抽出兩個球排列的事件共有幾種?

```python
from itertools import permutations 

# 測試資料
list1 = [1, 2, 3]

# 宣告排列的類別，三個球抽出兩個球
perm = permutations(list1, 2) 

# 印出所有事件
print('所有事件:')
list_output = list(perm)
for i in list_output: 
    print(i) 
    
print() 
print(f'排列數={len(list_output)}')
```

## 範例3. 有三個球，抽出兩個球組合的事件共有幾種?

```python
from itertools import combinations 

# 測試資料
list1 = [1, 2, 3]

# 宣告排列的類別，三個球抽出兩個球
comb = combinations(list1, 2) 

# 印出所有事件
print('所有事件:')
list_output = list(comb)
for i in list_output: 
    print(i) 
    
print() 
print(f'組合數={len(list_output)}')
```

## 範例4. 袋子中有10個不同顏色的球，抽出3個有哪幾種排列方式?

```python
import math

n=10
k=3

# math.factorial: 階乘
perm = math.factorial(n) / math.factorial(n-k)

print(f'排列方式共 {perm:.0f} 種')
```

## 範例5. 擲硬幣3次，出現0、1、2、3次正面的組合共有幾種?

```python
import math

n=3

# math.factorial: 階乘
for k in range(4):
    # 組合公式
    comb = math.factorial(n) / (math.factorial(k) * math.factorial(n-k))
    print(f'出現 {k} 次正面的組合共 {comb:.0f} 種')
```

# 機率

## 範例6. 擲硬幣出現正面的機率=0.4，擲硬幣3次，出現0、1、2、3次正面的機率為何?

```python
import math

n=3
p=0.4

# math.factorial: 階乘
for k in range(4):
    # 組合公式
    comb = math.factorial(n) / (math.factorial(k) * math.factorial(n-k))
    prob = comb * (p**k) * ((1-p)**(n-k))
    print(f'出現 {k} 次正面的機率 = {prob:.4f}')
```

## 使用 SciPy 的統計模組

```python
# 使用 scipy comb 組合函數
from scipy import special as sps

n=3
p=0.4

for k in range(4):
    prob = sps.comb(n, k) * (p**k) * ((1-p)**(n-k))
    print(f'出現 {k} 次正面的機率 = {prob:.4f}')
```

```python
# 從39個號碼選5個號碼，總共有幾種
from scipy import special as sps

print(f'從39個號碼選5個號碼，總共有{int(sps.comb(39,5))}種')

# 頭獎號碼的個數，5個中獎號碼全中，不含非中獎號碼
first_price_count = int(sps.comb(5,5) * sps.comb(34,0))
print(f'頭獎號碼的個數: {first_price_count}')

# 頭獎號碼的個數，5個中獎號碼中4個，有1個非中獎號碼
second_price_count = int(sps.comb(5,4) * sps.comb(34,1))
print(f'二獎號碼的個數: {second_price_count}')

# 頭獎號碼的個數，5個中獎號碼中3個，有3個非中獎號碼
third_price_count = int(sps.comb(5,3) * sps.comb(34,2))
print(f'三獎號碼的個數: {third_price_count}')

# 頭獎號碼的個數，5個中獎號碼中2個，有3個非中獎號碼
fourth_price_count = int(sps.comb(5,2) * sps.comb(34,3))
print(f'四獎號碼的個數: {fourth_price_count}')
```
